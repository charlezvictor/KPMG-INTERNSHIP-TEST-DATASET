### TRYING TO PREDICT LIST PRICE

```python
link = data_copy.drop(columns= ['transaction_id','product_id','customer_id', 'transaction_date', 'Type of Product','product_first_sold_date','order_status','Profit', 'Gender','Age','Job_Industry','Unnamed: 25','Deceased',], axis=1)
```

```python
link.head()
```

```python
# Create dummy variables for multiple columns
cols_to_transform = ['brand', 'product_line', 'product_class', 'product_size', 'Wealth Segment', 'Owns Car', 'Tenure', 'State']
link_dummy = pd.get_dummies(link[cols_to_transform])

# Concatenate the dummy variables with the original DataFrame
link1 = pd.concat([link, link_dummy], axis=1)
```

```python
link1.head()
```

```python
link1.columns
```

```python
# 
link_learn = link1.drop(columns=['brand', 'product_line','product_class', 'product_size','Wealth Segment', 'Owns Car', 'Tenure', 'State','online_order'],axis=1)
```

```python
link_learn.head()
```

```python
link_learn.isnull().sum()
```

```python
X = link_learn.drop(['list_price'],axis=1).values
y =link_learn['list_price'].values
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
reg.score(X_test, y_test)
```
